# **Descripción**



La pregunta central de la problemática a resolver es ¿Cómo ayudar a los alumnos de 15 años a mejorar sus habilidades y conocimientos en alguna de las 3 áreas de enfoque que evalúa la prueba PISA: lectura, matemáticas y ciencias?

PISA son las siglas de Programme for International Student Assessment, es decir, Programa para la Evaluación Internacional de Alumnos. La prueba PISA es un examen de la OCDE (Organización para la Cooperación y el Desarrollo Económicos) que se aplica cada 3 años y tiene como objetivo evaluar los conocimientos de los alumnos de 15 años una vez que culminan sus estudios obligatorios. La OCDE tiene 38 miembros alrededor del mundo; México es uno de los miembros actuales de dicha organización. 

En la prueba del año 2018, los estudiantes mexicanos obtuvieron un puntaje por debajo del promedio de los países que pertenecen a la OCDE en las áreas de lectura, matemáticas y ciencias. Además, se posicionó  en el tercer lugar en los países de America Latina, abajo de Chile y Uruguay. Los últimos resultados publicados demuestran una escasez de calidad en la educación del país y demandan que se implementen mejoras en el sistema educativo. 

Como propuesta de solución a la actual problemática se tiene lo siguiente: Desarrollar un programa computacional que les permita a los alumnos repasar las áreas de oportunidad que tengan, reforzar los conocimientos y obtener mejores resultados en la prueba PISA, así como mejorar sus habilidades y prepararse para la aplicación real de la prueba. El desarrollo del programa podrá permitir a México tener una base de datos de los resultados de los alumnos e identificar áreas en las que se necesita más trabajo, previo a la aplicación de la prueba PISA. Además, los mismos datos pueden permitir al país desarrollar estrategias y/o implementar modificaciones al sistema educativo actual. 

El programa le permitirá a los estudiantes aprender de forma autodidácta y a base de juego en distintas modalidades. La primera modalidad es estudiar temas dividos en las tres distintas áreas: lectura, matemáticas y ciencias, de esta manera se enfocan en un área a la vez. Por otro lado, los alumnos podrán presentar quizzes de las tres diferentes áreas como si fuera una simulación de la prueba PISA. 

El programa le dará la oportunidad a los estudiantes de conocer su avance en las distintas áreas y les brindará tips de técnicas de estudio para que mejoren sus resultados. 

# **Explicación de algoritmos**

**Algoritmo 1.** Alta de preguntas
La función tiene el objetivo de registrar las preguntas que le de el usuario y agregarlas a una lista de listas (matriz). La función tiene un ciclo while infinito que pregunta por el enfoque de la pregunta, si el enfoque el válido continua preguntando los demás datos, si no lo es vuelve a preguntar pro el enfoque.
Cuando el usuario termina de poner los datos de la pregunta que va a registrar la pregunta se agrega como una sublista a la lista principal <total_preguntas>. Si el usuario desea agregar otra pregunta se reinicia el ciclo, sino se despliega el menú principal.  

**Algoritmo 2.** Actualizar preguntas
Despliega la lista principal que contiene todas las preguntas. Inicia un ciclo while que pregunta por el número de pregunta que desea modificar el usuario. Valida si la pregunta existe; si no existe aparece error y reinicia el ciclo, si existe imprime la lista que pidió el usuario. Después pide al usuario los datos uno por uno y los almacena en una lista temporal. Una vez que termina de preguntar por los datos se pregunta al usuario si desea confirmar los cambios ingresados, si la respuesta es no la se imprime la matriz como se mostró al inicio y lo manda al menú principal. Si la respuesta es si, la lista nueva se pone en el lugar de la lista que solicitó el usuario y muestra como quedó la lista principal con las modificaciones y termina el ciclo. 

**Algoritmo 3.** Estudiar preguntas
La función solicita al usuario un área y una cantidad de preguntas. Inicia un ciclo for con range(len(cantidad de preguntas), para cada iteración hace una comparación; si el elemento 1 de cada una de las columnas es igual al enfoque que ingresó el usuario entonces se despliega la pregunta. 

**Algoritmo 4.** Presentar quiz
La función despliega 10 preguntas contenidas en la matriz principal. A través de un ciclo for que itera 10, si el elemento 1 de cada una de las columnas es igual a lectura se despliega la pregunta, si es igual a matemáticas se despliega la pregunta y si es igual a ciencias se despliega la pregunta. Las tres comparaciones anteriores son con if separados dentro del ciclo for.  

**Algoritmo 5.** Reporte de calificaciones
La función va a tener una matriz; una lista que almacenará un acumulador de cada vez que se pide esta función, una lista de promedio de calificaciones que va a sumar todos los resultados almacenados en quiz y estudiar preguntas y los va a dividir entre la cantidad de resultados para desplegar el promedio; preguntas correctas que será un contador de cuantas veces el mensaje “correcto” se desplegó y lista de preguntas incorrectas que contará cuantas veces el mensaje “incorrecto” se desplegó.

**Algoritmo 6.** Salir
La función salir únicamente despliega un mensaje al usuario de agradecimiento pro haber usado el sistema y para hacerle saber que su progreso y/ cambios fueron guardados.

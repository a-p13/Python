```python
from sklearn.linear_model import LinearRegression as lr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("peso_estatura_genero.csv")
x = df["Peso"].to_numpy().reshape((-1, 1))
y = df["Estatura"].to_numpy()
```

```python
sns.scatterplot(data=df, x="Peso", y = "Estatura", hue = "Genero")
```

```python
model = lr()  #inicializa un modelo de regresion lineal
```

```python
model.fit(x, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y2 = model.predict(x)
```

```python
model.score(x, y2)  #1 significa que esta bien
model.score(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, y2, "r")
plt.show()
```

# Problema 1. 
Escribe una función desempeño_alumno( ) que pregunte el nombre de un alumno y el número de semestre que cursa. La función deberá también preguntar al alumno las calificaciones de sus 5 materias del semestre actual para calcular e imprimir el promedio sin decimales. Recuerda que 0.5 o más redondea al entero siguiente.

La función deberá validar la etapa de estudio del alumno:

Si el alumno es de 1ro o 2do o 3er semestre, la función debe imprimir: El alumno "nombre" se encuentra en etapa de Exploración y su promedio en el semestre # es de:
Si el alumno es de 4to o 5to semestre, la función debe imprimir:  El alumno "nombre" se encuentra en etapa de Enfoque y su promedio en el semestre # es de:
Si el alumno es de 6to a 8vo semestre, la función debe imprimir:  El alumno "nombre" se encuentra en etapa de Especialización y su promedio en el semestre # es de:

```python
def desempeño_alumno():
  nombre = input("Ingresa tu nombre y apellido: ")
  semestre = int(input("Ingresa el número de semestre que cursas: "))
  suma = 0

  for calificaciones in range(1, 6):
    materia = float(input("Ingresa la calificación de tu materia " + str(calificaciones) + ": ")) 
    suma = suma + materia
  
  promedio = round(suma/5)

  if semestre == 1 or semestre == 2 or semestre == 3:
    print("El alumno ", nombre, "se encuentra en etapa de Exploración y su promedio en el semestre", semestre, " es de ", promedio)

  elif semestre == 4 or semestre == 5:
    print("El alumno ", nombre, "se encuentra en etapa de Enfoque y su promedio en el semestre", semestre, " es de ", promedio)

  elif semestre == 6 or semestre == 7 or semestre == 8:
    print("El alumno ", nombre, "se encuentra en etapa de Especialización y su promedio en el semestre", semestre, " es de ", promedio)

  else:
    print("El semestre que ingresaste no es válido.")

desempeño_alumno()
```

# Problema 2. 
Escribe una función comprar( ) que reciba como parámetro el tipo de pago (puede ser Efectivo o TDC) y que regrese el total de la compra. La función debe leer la clave del artículo que va a comprar (nota que es letra mayúscula) o X si ya no quiere comprar más artículos. La función debe mostrar en la pantalla el precio correspondiente a cada artículo pedido, y seguir preguntando mientras el usuario no teclee la clave X, cuando el usuario teclee la clave X el programa debe terminar y regresar el total de la compra del cliente.

```python
pago = input("Indica cual es tu tipo de pago (efectivo/TDC): ")

def comprar(pago):
  articulo = ""
  total = 0

  print("""
    A. Artículo $120
    B. Artículo $250 
    C. Artículo $360 
    X. Terminar compra""")

  while True:

    articulo = input("Indica la clave del artículo que vas a comprar: ")
  
    if articulo == "A":
      print("Precio $120")
      total = total + 120

    elif articulo == "B":
      print("Precio $250")
      total = total + 250

    elif articulo == "C":
      print("Precio $360")
      total = total + 360

    elif articulo == "X":
      print("Total a pagar en ", pago, "$")
      return total
      break

    else: 
      print("El artículo indicado no existe. Intenta de nuevo.")

comprar(pago)
```

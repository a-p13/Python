### Problema 1. 
Escribe una función que reciba 3 números enteros, que representan la longitud de los lados de un triángulo, y regresa el tipo de triángulo del que se trata (equilátero, isósceles o escaleno).

Considera que X, Y y Z son los lados de un triángulo, si cumplen con las siguientes condiciones:

Todos los números son mayores que cero.
La suma de dos de las medidas debe ser estrictamente mayor que la tercera:
X + Y > Z
X + Z > Y
Y + Z > X
Recuerda que el triángulo equilátero tiene 3 lados iguales, el isósceles tiene 2 lados iguales y el escaleno tiene los 3 lados diferentes.



```python
lado_x = int (input("Ingresa el valor del lado x "))
lado_y = int (input("Ingresa el valor del lado y "))
lado_z = int (input("Ingresa el valor del lado z "))

def tipo_triangulo (x, y, z): #recibe valores y te dice que tipo de triángulo es

  suma_1 = lado_x + lado_y
  suma_2 = lado_x + lado_z
  suma_3 = lado_y + lado_z

  if lado_x > 0 and lado_y > 0 and lado_z > 0:

    if  suma_1 > lado_z or suma_2 > lado_y or suma_3 > lado_x:

      if lado_x == lado_y and lado_y == lado_z:
        print("Es un triángulo equilatero.")

      elif lado_x == lado_y or lado_x == lado_z or lado_z == lado_y:
        print("Es un triángulo isósceles.")

      else:
        print("Es un triángulo escaleno.")

  else:
    print("Los valores no son válidos; no pueden hacer un triángulo.") 
  
tipo_triangulo(lado_x, lado_y, lado_z)    
```

# Problema 1. 
Diseña y codifica un programa en Python en el cual el usuario ingrese la cantidad de elementos que va a ingresar a la lista; posteriormente, el programa debe leer cada uno de los elementos de la lista, uno por línea y se van agregando a la lista.

Importante: El programa debe validar que el número de elementos a ingresar sea mayor que cero, sino debe volver a pedir el valor hasta que cumpla.

```python
lista = []

while True:
  elementos = int(input("Ingresa la cantidad de elementos a registrar: "))

  if elementos <= 0:
    print("Error. Debes ingresar un número mayor a 0.")
    continue

  else: 
    for elementos in range(1, elementos + 1):
      numero = int(input("Ingresa el elemento " + str(elementos) + ": "))
      lista.append(numero)
  
  break
      
promedio = sum(lista) / elementos

for indice in range(len(lista)):
  print("Lista [" + str(indice) + "] = " + str(lista[indice]))

print("El último número ingresado es ", lista[-1])
print("La suma de los elementos de la lista es ", sum(lista))
print("El promedio de los elementos ingresados es ", promedio)
   
```

# Problema 2. 
Escribe un programa que reciba del usuario una lista y devuelva otra con los elementos de la lista original, pero sin elementos duplicados.

```python
lista_usuarios = []

while True:
  elementos = int(input("Ingresa la cantidad de elementos a registrar: "))

  if elementos <= 0:
    print("Error. Debes ingresar un número mayor a 0.")
    continue

  else: 
    for elementos in range(1, elementos + 1):
      dato = (input("Ingresa el nombre " + str(elementos) + ": "))
      lista_usuarios.append(dato)
  
  break

print(lista_usuarios)

lista_nueva = []

for elemento in lista_usuarios:
  if elemento not in lista_nueva:
    lista_nueva.append(elemento)
  
print(lista_nueva)
```

# Problema 3. 
Desarrolla un programa que permitirá obtener, a partir de una lista que recibirá del usuario, crear y desplegar una sublista con valores pares y otra sublista con valores impares.

```python
lista_numeros = []
lista_pares = []
lista_impares = []
entero = 0

entero = input("Ingresa el elemento: ")

while entero != "*":

  lista_numeros.append(int(entero))

  entero = input("Ingresa el elemento: ")
  
for num in lista_numeros:
  if num % 2 == 0:
    lista_pares.append(num)
    
  else:
    lista_impares.append(num)

print("PARES")
print(lista_pares)
print("IMPARES")
print(lista_impares)    
```

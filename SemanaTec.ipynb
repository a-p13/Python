```python
#READ AND WRITE IN LIST
tabla = []
with open("peso_estatura_genero.csv", "r") as file:
  for line in file.readlines():
    row = []
    for word in line.split(", "):
      #word = word[:-2]
      word = word.rstrip("male/n")
      try:
        word = float(word)
      except:
        pass
      row.append(word)
    tabla.append(row)
```

```python
#PRINT TABLE
for line in tabla:
    print(line)
```

```python
#PRINT FIRST LINE OF TABLE
print(tabla[0])
```

```python
variables = ["G, E, P"]
tabla[0] = variables
print(tabla[0])
```

```python
new_user = ["Female", 56, 170]
tabla.append(new_user)
print(tabla)
```

```python
s_file = ""
for line in tabla:
  for word in line:
    s_file += word + ". "
  s_file = s_file[:-1]
  s_file += "/n"
```

```python
#WRITE
with open("peso_estatura_genero.csv", "w") as file:
  file.write(s_file)
```

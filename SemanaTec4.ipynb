```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
sns.countplot(data = df, y = "Species") #barras horizontales
#sns.countplot(data = df, x = "Species") #barras verticales
plt.title("Conteo de las especies")
plt.xlabel("Especies")
```

```python
#histograma
#sns.histplot(data = df, x = "Sepal.Length",  bins = 20)  #bins es numero de barras
#sns.histplot(data = df, x = "Sepal.Length", hue = "Species")  #Agrupa por especies
sns.histplot(data = df, x = "Sepal.Length", hue = "Species", kde = True, bins = 20)
```

```python
#Graficas de dispersion
#sns.scatterplot(data = df, x ="Sepal.Length", y = "Sepal.Width")
sns.scatterplot(data = df, x ="Sepal.Length", y = "Sepal.Width", hue = "Species")
```

```python
sns.pairplot(data = df) #grafico que junta todo
sns.pairplot(data = df, hue = "Species") #grafico que junta todo agrupado por especies
```

```python
df_corr = df.corr()
```

```python
sns.heatmap(data = df_corr, vmin = -1, vmax = 1)
sns.heatmap(data = df_corr, vmin = -1, vmax = 1, cmap = "cividis", annot = True)
```

```python
sns.boxplot(data=df, x="Sepal.Length", y = "Species")
```

### Problema 3. Haz una función sirva para calcular el precio que va a pagar un cliente por comprar cemento. La función debe leer la cantidad de bultos de cemento que va a comprar el cliente, y el precio del bulto de cemento. La función debe mostrar como salida 3 datos uno en cada renglón: el precio antes de impuestos, los impuestos (que son el 16% del precio) y el total a pagar por el cliente.



```python
def calculo_precio (): #definimos la función llamada precio

  cantidad_bultos = int(input("Ingresa la cantidad de bultos que comprará el cliente: ")) #bultos enteros
  precio_bulto = float(input("Ingresa el precio por bulto del cemento: ")) #precio en decimal

  subtotal = cantidad_bultos * precio_bulto
  iva = subtotal * 0.16
  total = subtotal + iva 

  print("Subtotal $", subtotal)
  print("IVA $", iva)
  print("Total $", total)

calculo_precio() #llama la función calculo_precio
```

### Problema 4. A inicio de año, las personas están preocupadas por su peso, por lo que acuden a nutriólogos, gimnasios y cualquier otra cosa que les ayude en el proceso. Realiza una función que ayude a las personas a indicar cuántos kilos debe bajar por mes dados el peso inicial, el peso final y el número de meses que una persona estará en un programa integral para bajar de peso. Solicitudes (entradas): Peso inicial (número decimal), peso final (número decimal), meses (número entero). Un dato en cada línea en ese orden.

```python
def plan ():
  peso_actual = float(input("Ingresa tu peso actual en kg ")) 
  peso_deseado = float(input("Ingresa tu peso deseado en kg ")) 
  meses = int(input("¿En cuántos meses quieres llegar al peso deseado? "))

  kilos = peso_actual - peso_deseado 
  kilos_por_mes = kilos / meses

  print("La cantidad de kg a bajar es de ", kilos)
  print("Al mes tendrás que bajar ", kilos_por_mes, "kg, para lograr tu objetivo") 

plan()
```

### Problema 6. Desarrolla una función para calcular las siguientes medidas de un círculo/esfera/cilindro recibiendo el radio = 3cm y altura = 10
1. Diámetro

2. Perímetro o circunferencia

2. Área

3. Volumen de la esfera

4. Volumen de un cilindro


```python
import math
radio = 3 #cm
altura = 10 # cm

def medidas (radio, altura):
  diametro = 2*radio
  perimetro = 2*radio*math.pi
  area = math.pi * (radio**2)
  volumen_esfera = 4/3 * math.pi * (radio**3)
  volumen_cilindro = math.pi * (radio**2) * altura

  print("El diámetro del círculo es ", diametro, "cm")
  print("El perímetro del círculo es ", perimetro, "cm")
  print("El área del círculo es ", area, "cm^2")
  print("El volumen de la esfera es ", volumen_esfera, "cm^3")
  print("El volumen del cilindro es ", volumen_cilindro, "cm^3")

medidas (radio, altura)
```

### Problema 7. Realiza una función para calcular el área de un triángulo cuyos lados son de longitud a, b y c.

```python
import math
a = int(input("Ingresa el valor del lado a "))
b = int(input("Ingresa el valor del lado b "))
c = int(input("Ingresa el valor del lado c ")) 

def calculo_area_triangulo (a, b, c):
   
  s = (a+b+c)/2
  area = math.sqrt(s*(s-a)*(s-b)*(s-c))

  print ("El área del triángulo es ", area)

calculo_area_triangulo(a, b, c)
```

### Problema 8. Escribe una función para calcular la distancia entre dos puntos del plano cartesiano solicitando el punto inicial y el final en términos de x y y, para desplegar el valor de la distancia.

```python
import math
def calculo_distancia ():
  x1 = int(input("Ingresa el valor inicial de x: "))
  y1 = int(input("Ingresa el valor inicial de y: "))
  x2 = int(input("Ingresa el valor final de x: "))
  y2 = int(input("Ingresa el valor final de y: "))

  distancia = math.sqrt(((x2-x1)**2)+((y2-y1)**2))
  print("La distancia entre los puntos ingresados en el plano cartesiano es de ", distancia)

calculo_distancia()
```

Escribe una función que reciba un número positivo n, e imprima todos los números en orden desde el 1 hasta n. Cada uno de los números debe ser impreso en una linea por separado.

```python
numero = int(input("ingresa un nummero"))


def numeros(numero):
  
  if numero >= 1:
    for num in range(1, numero+1):
      print(num) 

numeros(numero)
```

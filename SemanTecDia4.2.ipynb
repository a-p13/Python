```python
from sklearn.linear_model import LinearRegression as lr
from sklearn.preprocessing import PolynomialFeatures as pf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("peso_estatura_genero.csv")
```

```python
polynomial = pf(degree=2, include_bias = False)
```

```python
x =df["Peso"].to_numpy().reshape((-1, 1))
y =df["Estatura"].to_numpy()
```

```python
polynomial_x = polynomial.fit_transform(x)  #añadir variables y transformar para obtener caracteristicas
```

```python
model_linear = lr()
model_linear.fit(polynomial_x, y)
```

```python
y2 = model_linear.predict(polynomial_x)
```

```python
model_linear.score(polynomial_x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, y2, "r")
plt.show()
```

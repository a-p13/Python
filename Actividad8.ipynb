# Problema 1. 
Muestra el promedio de todos los números capturados por el usuario. Primero, pregunta por la cantidad de números que el usuario desea capturar y luego pregunta por cada uno de los números y después despliega en pantalla el promedio de todos los números

```python
repeticion = int(input("Ingresa la cantidad de números que deseas capturar: "))
suma = 0
  
for num in range (1, repeticion + 1):
  numero = int(input("Ingresa el número: " ))
  suma = suma + numero
  
promedio = suma / repeticion
print("El promedio de los números que ingresaste es de ", promedio) 
```

# Problema 4. 
Suma los números enteros (positivos y negativos) que el usuario teclee, deberás dejar de preguntar por números hasta que la suma de todos los números capturados sea igual a cero

```python
numero = int(input("Ingresa un número entero (positivo o negativo): "))
suma = 0
suma = suma + numero

while suma != 0:
  numero = int(input("Ingresa un número entero (positivo o negativo): "))
  suma = suma + numero
  print("La suma es de: ", suma)

print("La suma de los números que diste es cero") 
```

# Problema 5.
Recupera el código de tu programa de la calculadora de IMC (índice de masa corporal) y realiza las adecuaciones para que la calculadora le permita al usuario calcular repetidamente el IMC de los pacientes en un consultorio.

```python
while True:
  peso = float(input("Ingresa el peso de en kg "))
  altura = float(input("Ingresa la altura en metros "))

  imc = peso / (altura ** 2)

  print("El Índice de Masa Corporal (IMC) es de, ", round(imc, 2))
  print("El IMC indica que está en el siguiente rango: ")

  if imc < 20: 
    print("Peso bajo")

  elif imc < 25 and imc >= 20: 
    print("Peso normal")

  elif imc < 30 and imc >= 25:
    print("Sobrepeso")

  elif imc < 40 and imc >= 30:
    print("Obesidad")

  else:
    print("Fuera de rango")

  continuar = input("¿Deseas calcular el IMC de otro paciente? (S/N)")

  if continuar == "S" or continuar == "s" or continuar == "Si" or continuar == "si" or continuar == "Sí" or continuar == "sí":
    continue

  else:
    print("Gracias por usar la calculadora.")
    break
```

# Problema 6. 
Escribe un programa que pregunte el nombre del usuario y contraseña, que valide si son correctas y si alguna no es correcta, que vuela a preguntar hasta que sean correctas y despliegue el mensaje: "Bienvenido, usuario."

```python
nombre = "Andrea"
contraseña = "123"
contador = 0

while True:
  nombre_usuario = input("Ingresa el nombre de usuario: ")
  contraseña_usuario = input("Ingresa la contraseña: ")
  
  if nombre == nombre_usuario and contraseña == contraseña_usuario:
    print("Bienvenido,", nombre)
    break
  
  else:
    print("Nombre o contraseña incorrecto. Intenta de nuevo.")
    contador = contador + 1
  
  if contador == 3:
    print("Límite de intentos excedido.")
    break
```

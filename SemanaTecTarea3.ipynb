```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DA_Hospitalizados.csv")
```

```python
#Mostrar numero de columnas y numero de filas
df.info()
```

```python
df.head()
```

```python
sns.countplot(data = df, y = "Institución") #barras horizontales
#sns.countplot(data = df, x = "Species") #barras verticales
plt.title("Pacientes que entran a cada Institucion")
plt.xlabel("Pacientes")
```

```python
sns.pairplot(data = df) #grafico que junta todo
sns.pairplot(data = df, hue = "Positivos") #grafico que junta todo agrupado por especies
```

```python
df_corr = df.corr()
```

```python
sns.heatmap(data = df_corr, vmin = -1, vmax = 1)
sns.heatmap(data = df_corr, vmin = -1, vmax = 1, cmap = "cividis", annot = True)
```

```python
plt.figure(figsize = (20,5))
inst = df[(df["Institución"] == "SSJ") | (df["Institución"] == "HCG") | (df["Institución"] =="ISSSTE") | (df["Institución"] == "Privados") | (df["Institución"] == "IMSS")]

sns.boxplot(data=inst, x="Positivos", y = "Institución")
```

```python
plt.figure(figsize = (20,5))
inst2 = df[(df["Institución"] == "SMM TLAJOMULCO") | (df["Institución"] == "SMM GUADALAJARA") | (df["Institución"] =="SSMM ZAPOPAN") | (df["Institución"] == "CRUZ ROJA")]
inst2_filtrada = inst2[inst2["Pacientes"].notnull()]
sns.boxplot(data=inst2_filtrada, x="Pacientes", y = "Institución")
```

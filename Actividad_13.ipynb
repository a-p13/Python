# Problema 1. 
Crea dos matrices 'A' y 'B' de 3 X 3 con datos del 1 al 9 e imprime sus listas y valores renglón por renglón.

```python
A = [
     [1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]
]

B = [
     [1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]
]
```

```python
print("Lista A.", A)
for lista in A:
  for _ in lista:
    print(_)

print("Lista B.", B)
for lista in B:
  for _ in lista:
    print(_)
```

# Problema 3. 
Realiza un programa que cree una función que reciba la matriz A y sume todos sus elementos e imprima el total.

```python
def matriz(A):
  suma = 0

  for lista in A:
    for _ in lista:
      suma = suma + _

  print("La suma de los elementos de la matriz es: ", suma)

matriz(A)
```

# Problema 4. 
Realiza un programa que cree una matriz de n filas y m columnas que deberá estar llena de números consecutivos por renglón, comenzando por 1.

```python
while True:

  filas = int(input("Ingresa el número de filas: "))
  columnas = int(input("Ingresa el número de columnas: "))

  if filas < 2 or columnas < 2:
    print("Error.")
    continue   

  else:
    contador = 1
    matriz = []

    for fila in range(filas):
      lista = []
        
      for columna in range(columnas):
        lista.append(contador)
        contador = contador + 1

      matriz.append(lista)

    for renglon in matriz:
      print(renglon)
    break
```

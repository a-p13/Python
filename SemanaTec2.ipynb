```python
#lib / 2.2
#inches *2.54
dt.Estatura = dt.Estatura*2.54
dt.Peso = dt.Peso/2.2
```

```python
print(dt)
```

```python
print(dt.columns)
```

```python
print(dt.describe())
```

```python
#imprime valores donde la estatura sea mayor a 170
dt_170  = dt[dt["Estatura"]>170]
print(dt_170)
```

```python
#imprime valores donde la estatura sea mayor a 170
dt_170  = dt[(dt["Genero"]=="Female") & ((dt["Peso"]<150) | (dt["Estatura"]<170))]
print(dt_170)
print(dt_170.describe())
```

```python
#promedio de la estatura
media = dt["Estatura"].mean()
print(media)
```

```python
#promedio con filtrado
dt_female  = dt[(dt["Genero"]=="Female")]
media = dt_female.mean()
print(media)
```

```python
#te da el min y maximo de todo
dt.min()
dt.max()
```

```python
#media recortada
estatura_sorted = dt["Estatura"].sort_values().reset_index(drop = True)

p = 100
total = 0
for i in range(p, len(estatura_sorted)-p):
  total += estatura_sorted[i]
  rmean = total/len(estatura_sorted-2*p)
  print(rmean)
```
